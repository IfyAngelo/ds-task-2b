```python
# Function to clean and normalize the first name
def clean_and_normalize_name(name):
    # Remove leading and trailing whitespaces
    cleaned_name = name.strip()
    # Convert the name to lowercase
    normalized_name = cleaned_name.lower()
    return normalized_name

# Apply the cleaning function to the CONTACTFIRSTNAME column
sales_data['CONTACTFIRSTNAME'] = sales_data['CONTACTFIRSTNAME'].apply(clean_and_normalize_name)
```

```python
# Function to map all unique names to male gender
def map_names_to_male(names):
    # Convert names to lowercase for case-insensitive matching
    names_lower = [name.lower() for name in names]
    # Create a dictionary mapping each unique name to the male gender
    name_gender_mapping = {name: "male" for name in set(names_lower)}
    return name_gender_mapping

# Get unique names from the CONTACTFIRSTNAME column
unique_names = sales_data['CONTACTFIRSTNAME'].unique()

# Map all unique names to male gender
name_gender_mapping = map_names_to_male(unique_names)

# Display the name-gender mapping
print(name_gender_mapping)
```
